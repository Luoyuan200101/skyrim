```python
@dataclass(slots=True)
class ModelPrediction:
    filepath: str
    prediction: xr.Dataset | xr.DataArray 

    def __init__(self, source):
        if isinstance(source, str):
            self.filepath = source
            self.prediction = xr.open_dataarray(source).squeeze()
        
        elif isinstance(source, xr.Dataset) or isinstance(source, xr.DataArray):
            self.filepath = None
            self.prediction = source

    @property
    def coords(self):
        return list(self.prediction.coords)

    @property
    def size(self):
        return self.prediction.sizes

    @property
    def variables(self):
        return self.prediction.channel.values
```
